```python
def train_test_finder(i):
    """
    Returns 0 if the index is in test data, and 1 if it is in train data
    i : int value of an index that is part of the redwine dataframe
    """
    return int(i in train_data.index.tolist())
    
```

```python
train_test_finder(1328)
```

```python
redwine.loc[:,"train"] = pd.Series(redwine.index.tolist()).apply(train_test_finder)
```

```python
redwine
```

```python
g = sns.FacetGrid(data=redwine,col="train")
g.map(sns.barplot,"quality","alcohol")
plt.show()
```

```python
feature_cols = [
    'fixed acidity',
    'volatile acidity',
    'citric acid',
    'residual sugar',
    'chlorides',
    'free sulfur dioxide',
    'total sulfur dioxide',
    'density',
    'pH',
    'sulphates',
    'alcohol'
]
```

```python
target_col = "quality"
```

```python
X_train = train_data[feature_cols].values

Y_train = train_data[target_col].values
```

```python
X_train
```

```python
Y_train
```

```python
X_test = test_data[feature_cols].values

Y_test = test_data[target_col].values
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn import tree
```

```python
# initialize Decision Tree classifier
model = tree.DecisionTreeClassifier()
```

```python
# train model on train data
model.fit(X_train,Y_train)
```

```python
# predict on test data
Y_test_predicted = model.predict(X_test)
```

```python
Y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(Y_test,Y_test_predicted))
```

```python
train_data[train_data.quality == 3]
```

```python
redwine.quality.unique()
```

```python
cm = pd.DataFrame(
    confusion_matrix(
        Y_test,
        Y_test_predicted,
        labels=[3,4,5,6,7,8]
    ),
    columns=[3,4,5,6,7,8],
    index = [3,4,5,6,7,8]
)


cm["total"] = cm.sum(axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# initialize Random Forest classifier
rf = RandomForestClassifier()
```
