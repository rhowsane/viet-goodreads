```python
import pandas as pd
import csv

# Load the dataset
pd.set_option("display.max_colwidth", None)
df_from_csv = pd.read_csv('Dataset/books_0.2.csv')
df_from_csv.head(3)
```

```python
df_filtered_rows = df_from_csv.loc[(df_from_csv["column1"].isna())&(df_from_csv["column2"].isna())&(df_from_csv["column3"].isna())]
df_filtered_rows_and_columns = df_filtered_rows.iloc[:,0:-3]
df_filtered_rows_and_columns.head()
```

```python
df_types = df_filtered_rows_and_columns.dtypes.reset_index().rename(columns={0:"dtype"})
df_types
```

```python
categorical_columns = []
numerical_columns = []

for column, type in zip(df_types["index"],df_types["dtype"]):
    if type == "object":
        categorical_columns.append(column)
    else:
        numerical_columns.append(column)

len(categorical_columns), len(numerical_columns)
```
